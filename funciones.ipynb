```python
import numpy as np  

def f(x):
    return np.sqrt(x + 2)
```

```python
x = np.array([-2, -1, 0, 2, 4, 6])  # Creando el vector de valores de x
y = f(x)
y
```

```python
import pandas as pd

tabla = pd.DataFrame( list(zip(x, y)), columns=['x', 'f(x)'] )
tabla
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

def move_spines():
    """Esta funcion divide pone al eje y en el valor 
    0 de x para dividir claramente los valores positivos y
    negativos."""
    fix, ax = plt.subplots()
    for spine in ["left", "bottom"]:
        ax.spines[spine].set_position("zero")
    
    for spine in ["right", "top"]:
        ax.spines[spine].set_color("none")
    
    return ax

x = np.linspace(-2, 6, num=30)

ax = move_spines()
ax.grid()
ax.plot(x, f(x))
plt.title(r"Grafico de $f(x)=\sqrt{x + 2}$")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
datas = np.linspace(0,2,10000)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(datas, datas, label= "linear")
ax1.plot(datas, datas**2, label = "cuadratic")
ax2.plot(datas, datas**3, label = "cubic")
ax2.plot(datas, np.sqrt(datas), label = "sqrt")
ax1.set_xlabel("Variable independiente")
ax1.set_ylabel("Variable dependiente")
ax1.set_title("Gráfico inicial")
ax1.legend()
ax2.legend()
```

```python
#Mismo ejercicio de antes, pero incorporando toda la información del gráfico en una línea, y detallando las características de la línea

datas = np.linspace(0,2,10000)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(datas, datas, "r--", datas, datas**2, "bs")
ax2.plot(datas, datas**3, "c,-", datas, np.sqrt(datas), "gh:")
ax1.set_xlabel("Dep")
ax1.set_ylabel("Ind")
ax2.set_xlabel("Dep")
ax2.set_ylabel("Ind")
```
